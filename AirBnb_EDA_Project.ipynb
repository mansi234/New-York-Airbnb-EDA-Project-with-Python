# **Number of reviews and price relation*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
data=pd.read_csv('datasets.csv', encoding_errors ='ignore')
df=data[data['price']<1500]
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
sns.pairplot(data=df,vars=['minimum_nights','availability_365','number_of_reviews','price'],hue = 'room_type')
```

# **Geographical distribution of AirBnb listings*

```python
sns.scatterplot(data=df ,x='longitude', y='latitude', hue='room_type')
plt.show()
```

# heatmap - Correlation of one variable with other for numerical columns



```python
corr =df[['longitude', 'latitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']].corr()
corr
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(data=corr, annot = True)
```
